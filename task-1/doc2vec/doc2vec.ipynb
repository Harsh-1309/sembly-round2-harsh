```python
import gensim
import pandas as pd
import numpy as np
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import nltk
```

```python
df=pd.read_csv("task-1/data/cleaned_data.csv")
df.info()
```

```python
all_questions = list(df['question1'][:40000]) + list(df['question2'][:40000])
type(all_questions)
```

```python
nltk.download("punkt")
```

```python
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(all_questions)]
```

```python
tagged_data
```

```python
max_epochs = 100
vec_size = 200
alpha = 0.025
```

```python
model = Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                dm =1)
```

```python
# model = gensim.models.doc2vec.Doc2Vec(vector_size=40, min_count=2, epochs=30)
```

```python
model.build_vocab(tagged_data)
```

```python
from tqdm import tqdm
```

```python
for epoch in tqdm(range(max_epochs)):
    # print('iteration {0}'.format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha
```

```python
model.save("task-1/data/doc2vec/d2v.model")
```
