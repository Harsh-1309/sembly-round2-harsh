```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data="task-2/data/"
```

```python
train_data = pd.read_csv(data+"train.csv")
spray_data = pd.read_csv(data+"spray.csv")
weather_data = pd.read_csv(data+"weather.csv")
```

```python
spray_data.isnull().any()
```

```python
weather_data=weather_data.replace(["-","M"],np.nan)
weather_data=weather_data.replace(['  T'],0.00)
```

```python
weather_data.isnull().any()
```

```python
weather_data["DewPoint"].unique()
```

```python
weather_data.drop(["Tavg","Sunrise","Sunset","Water1","SnowFall","Heat","Cool","Depart","CodeSum","Depth","StnPressure"],axis=1,inplace=True)
```

```python
weather_data.isnull().sum()
```

```python
station_1_split=weather_data[weather_data.Station==1]
station_2_split=weather_data[weather_data.Station==2]
```

```python
station_1_split.isnull().sum()
```

```python
station_2_split.isnull().sum()
```

```python
def fill_st1_from_st2(cols):
    for col in cols:
        na_val_index=station_1_split[station_1_split[col].isnull()].index.to_list()
        for i in na_val_index:
            station_1_split[col][i]=station_2_split[col][i+1]
```

```python
def fill_st2_from_st1(cols):
    for col in cols:
        na_val_index=station_2_split[station_2_split[col].isnull()].index.to_list()
        for i in na_val_index:
            station_2_split[col][i]=station_1_split[col][i-1]
```

```python
fill_st1_from_st2(['WetBulb','SeaLevel',"PrecipTotal"])
```

```python
fill_st2_from_st1(['AvgSpeed','WetBulb','PrecipTotal','SeaLevel'])
```

```python
station_1_split
```

```python
station_2_split
```

```python
station_merged=pd.concat([station_1_split, station_2_split], join="inner")
```

```python
station_merged
```

```python
cleaned_train_data=pd.merge(train_data,station_merged,on="Date")
```

```python
cleaned_train_data[['Year', 'Month', "Day"]] = cleaned_train_data['Date'].str.split('-', 0, expand=True)
```

```python
cleaned_train_data.drop(columns=["Address", "AddressNumberAndStreet","Date","Station"],inplace=True)
```

```python
cleaned_train_data.info()
```

```python
lbl = LabelEncoder()
lbl.fit(list(cleaned_train_data['Species'].values))
cleaned_train_data['Species'] = lbl.transform(cleaned_train_data['Species'].values)

lbl.fit(list(cleaned_train_data['Street'].values))
cleaned_train_data['Street'] = lbl.transform(cleaned_train_data['Street'].values)

lbl.fit(list(cleaned_train_data['Trap'].values))
cleaned_train_data['Trap'] = lbl.transform(cleaned_train_data['Trap'].values)
```

```python
cleaned_train_data['Day'] = cleaned_train_data['Day'].astype(int)
cleaned_train_data['Month'] = cleaned_train_data['Month'].astype(int)
cleaned_train_data['Year'] = cleaned_train_data['Year'].astype(int)
cleaned_train_data['WetBulb'] = cleaned_train_data['WetBulb'].astype(int)
cleaned_train_data['PrecipTotal'] = cleaned_train_data['PrecipTotal'].astype(float)
cleaned_train_data['SeaLevel'] = cleaned_train_data['SeaLevel'].astype(float)
```

```python
cleaned_train_data.head()
```

```python
cleaned_train_data.to_csv(data+"cleaned_data.csv",index=False)
```
